```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
ruta=("/gdrive/MyDrive/CURSOS-ESPECIALIZACION/ML_Python-DMC/avance/data/Datos.csv")
```

```python
data=pd.read_csv(ruta)
```

```python
data.head(2)
```

```python
data=pd.DataFrame(data)
data.head(1)
```

```python
### Y=AX+B ###
x=data[["Valor1"]]
y=data[["Valor2"]]
plt.scatter(x,y)
plt.xlabel("valor numero 1")
plt.ylabel("valor numero 2")
plt.show()#buenas practicas se pone esto
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
modelo_rl=linear_model.LinearRegression()
modelo_rl.fit(x_train, y_train)#entrena
```

```python
print("precision del modelo:")
print(modelo_rl.score(x_train,y_train))
```

```python
y_pred=modelo_rl.predict(x_test)#va a comparar con y_test ¿para ver que tan efectivo es nuestro modelito?
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred,color="blue",linewidth=3)
plt.title("regresion lineal simple")
plt.show()
```

```python
print("ecuacion de la regresion")
print("y = ",modelo_rl.coef_,"x + ",modelo_rl.intercept_)
```
